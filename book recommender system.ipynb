### Popularity Based Recommendation System

```python
ratings_df.head(3)
```

```python
# merge the ratings with books to have a ratings that have book names
ratings_with_book_name = ratings_df.merge(book_df, on='ISBN')
```

```python
ratings_with_book_name.head(1)
```

```python
# try to create a dataframe for number of ratings for each book
number_of_ratings = ratings_with_book_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
number_of_ratings.head(3)
```

```python
number_of_ratings.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
```

```python
number_of_ratings.head(2)
```

```python
avg_ratings = ratings_with_book_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings.rename(columns={'Book-Rating':'avg_ratings'}, inplace = True)
```

```python
avg_ratings.head(2)
```

```python
popular_df = number_of_ratings.merge(avg_ratings, on='Book-Title')
popular_df
```

```python
# we wamt the books that have ratings of more than 250 and we want 50 0f those books
popular_df = popular_df[popular_df['num_ratings']>250].sort_values('avg_ratings', ascending=False).head(50)
```

### Collaborative Filtering Based Recommender System

```python
x_value = ratings_with_book_name.groupby('User-ID').count()['Book-Rating'] > 200 # get users that have booking rataing greater than 200
```

```python
educated_users = x_value[x_value].index
```

```python
# if the user id is in educated users then filter that user
filtered_ratings = ratings_with_book_name[ratings_with_book_name['User-ID'].isin(educated_users)]
```

```python
filtered_ratings.head(2)
```

```python
# to get the y value, we count the number of book titles and the ones greater tha 50 will be captured
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
famous_books = y[y].index
```

```python
final_filter = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_filter.head(2)
```

```python
pt = final_filter.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
# function to recommend books
def recommend(input_name):
    # get the inddex
    idx = np.where(pt.index == input_name)[0][0]
    # get the similar itemss
    similar_items = sorted(list(enumerate(similarity_scores[idx])),key= lambda x: x[1], reverse = True)[1:7]
    # list to store the data
    data = []
    # traverse the similar items andd get the items that similar with what the user requested
    for i in similar_items:
        items = []
        temp_df = book_df[book_df['Book-Title'] == pt.index[i[0]]]
        # add the items to the end of the list to create an iterable list
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(items)
    
    return data
```

```python
recommend('The Notebook')
```
